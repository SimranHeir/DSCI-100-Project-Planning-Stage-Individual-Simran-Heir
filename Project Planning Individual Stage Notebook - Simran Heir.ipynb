The data is about information about players that play video games, the duration of playing video games, the player email, and whether they would subscribe to a game related newsletter.

Session data:

There are 5 variables: hashEmail, start_time, end_time, original_start_time, and original_end_time.
There are 1535 observations for this dataset.

Reorganizing data

```r
organizing_start_data <- sessiondata |>
    separate (start_time, into = c("start_date", "start_time"), sep = " ")

organizing_end_data <- organizing_start_data |>
    separate (end_time, into = c("end_date", "end_time"), sep = " ")
```

Reorganizing original time column

Reorganize

```r
original_time_hours <- organizing_end_data |>
    mutate (original_start_time_hours = original_start_time/3600000, original_end_time_hours = original_end_time/3600000)
```

What to change the start time and end time to just average time, so that I can combine the data sets.

```r
session_data <- original_time_hours |>
    mutate (played_hours = (sqrt ((original_end_time-original_start_time)^2)))
```

```r
head (session_data, 6)
```

Combining the two data sets (player data + session data)

using semi_join to combine the player data and session data

```r
joined_data <- semi_join (player_data, session_data)
joined_data
```
